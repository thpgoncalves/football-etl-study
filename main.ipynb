```python
from statsbombpy import sb
```

```python
free_comps = sb.competitions()

free_comps
```

```python
free_comps_set = set(free_comps['country_name'])

free_comps_set
```

```python
free_comps_filtered = free_comps[(free_comps['country_name'] == 'Spain') & 
           (free_comps['competition_gender'] == 'male') &
           (free_comps['competition_name'] == 'La Liga')
           ].sort_values(by='season_name', ascending=True)

filtered_df = free_comps_filtered[free_comps['season_name'] != '1973/1974']
```

```python
competition_data = {
    f"{row['competition_name']} - {row['season_name']}" : {
        "competition_id": row['competition_id'],
        "season_id": row['season_id']
    }
    for _, row in filtered_df.iterrows()
}

for key, value in (competition_data.items())[:5]:
    print(key, ":", value)
```
